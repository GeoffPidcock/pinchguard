```python
import pandas as pd
import matplotlib.pyplot as plt

PROJECTIONS_DIR = "../assistant_probing/data/projections"

df_run00 = pd.read_csv(f"{PROJECTIONS_DIR}/scenario_07_treatment_run00/scenario_07_treatment_run00.csv")
df_run01 = pd.read_csv(f"{PROJECTIONS_DIR}/scenario_07_treatment_run01/scenario_07_treatment_run01.csv")
df_run02 = pd.read_csv(f"{PROJECTIONS_DIR}/scenario_07_treatment_run02/scenario_07_treatment_run02.csv")

df_run00.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(df_run00["turn"], df_run00["label_probe_L32"], marker="o")
axes[0].set_title("Run 00 - Layer 32")
axes[0].set_xlabel("turn")
axes[0].set_ylabel("label_probe_L32")

axes[1].plot(df_run00["turn"], df_run00["label_probe_L50"], marker="o", color="orange")
axes[1].set_title("Run 00 - Layer 50")
axes[1].set_xlabel("turn")
axes[1].set_ylabel("label_probe_L50")

fig.suptitle("Scenario 07 Treatment - Run 00")
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(df_run01["turn"], df_run01["label_probe_L32"], marker="o")
axes[0].set_title("Run 01 - Layer 32")
axes[0].set_xlabel("turn")
axes[0].set_ylabel("label_probe_L32")

axes[1].plot(df_run01["turn"], df_run01["label_probe_L50"], marker="o", color="orange")
axes[1].set_title("Run 01 - Layer 50")
axes[1].set_xlabel("turn")
axes[1].set_ylabel("label_probe_L50")

fig.suptitle("Scenario 07 Treatment - Run 01")
fig.tight_layout()
plt.show()
```
